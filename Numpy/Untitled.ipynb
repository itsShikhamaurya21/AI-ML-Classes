```python
arr3D=np.array([
    [
        [2,3,4],
        [5,6,4]
    ],
    [
        [3,4,5],
        [7,8,9]
    ]
])
```

```python
arr3D
```

```python
# axis 0 2+3 3+4 4+5 ----column wise on z axis
arr3D.sum(axis=0)
```

```python
# 2+5 3+6 4+4  ---column wise on y axis
arr3D.sum(axis=1)
```

```python
# 2+3+$  5+6 +4  -----row wise on x axis
arr3D.sum(axis=2)
```

```python
np.average(arr3D)
```

```python
np.max(arr3D)
```

```python
np.min(arr3D)
```

```python
np.min(arr3D,axis=0 )
```
