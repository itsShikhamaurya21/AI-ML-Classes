```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints=np.array([2,5])
ypoints=np.array([100,300])
plt.plot(xpoints,ypoints,color='green', linestyle='-.', marker='p')
plt.title("Custom line plot")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.grid(True)
plt.show()
```
